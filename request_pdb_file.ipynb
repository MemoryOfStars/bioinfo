```python
import pandas as pd
import requests
```

```python
pdb_id = pd.read_csv('DRAMP_Antimicrobial_amps.CSV',encoding='gb18030')
pdb_id.head()
```

```python
import numpy as np

pdb_rows = np.array(pdb_id.PDB_ID)
pdb_rows
```

```python
for i in pdb_rows:
    if i != 'Unknown':
        url = 'https://files.rcsb.org/download/' + i[0:4] + '.pdb'
        r=requests.get(url)
        with open('./pdb_file/'+i[0:4] + '.pdb','wb+') as pdb_file:
            pdb_file.write(r.content)
```
